```python
data = pd.read_csv('BTC17.csv')
```

```python
data.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
```

```python
WINDOW = 10
ROLLING = WINDOW / 2
STEP = 1
FORECAST = 1

highp = pd.to_numeric(data.ix[:, 'high'])
lowp = pd.to_numeric(data.ix[:, 'low'])
openp = pd.to_numeric(data.ix[:, 'open'])
closep = pd.to_numeric(data.ix[:, 'close'])
volume = pd.to_numeric(data.ix[:, 'volume'])

normal_close = closep
normal_open = openp
normal_close = np.array(normal_close)
normal_open = np.array(normal_open)

highp = highp.pct_change().replace(np.nan, 0).replace(np.inf, 0)
lowp = lowp.pct_change().replace(np.nan, 0).replace(np.inf, 0)
openp = openp.pct_change().replace(np.nan, 0).replace(np.inf, 0)
closep = closep.pct_change().replace(np.nan, 0).replace(np.inf, 0)
volume = volume.pct_change().replace(np.nan, 0).replace(np.inf, 0)

highp = np.clip(np.array(highp), -1, 1)
lowp = np.clip(np.array(lowp), -1, 1)
openp = np.clip(np.array(openp), -1, 1)
closep = np.clip(np.array(closep), -1, 1)
volume = np.clip(np.array(volume), -1, 1)
```

```python
def remap(x, out_min, out_max):
    in_min, in_max = np.min(x), np.max(x)
    return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
```

```python
from feature_calculators import *

rolling_volatility = pd.Series(closep).rolling(ROLLING).std()
rolling_second_derivative = pd.Series(closep).rolling(ROLLING).apply(mean_second_derivative_central)
rolling_absolute_energy = pd.Series(closep).rolling(ROLLING).apply(abs_energy)
```

```python
X, Y, MARGINS, BENCHMARK = [], [], [], []
for i in range(ROLLING, len(openp), STEP): 

    try:
        o = openp[i:i+WINDOW]
        h = highp[i:i+WINDOW]
        l = lowp[i:i+WINDOW]
        c = closep[i:i+WINDOW]
        v = volume[i:i+WINDOW]
        volat = rolling_volatility[i:i+WINDOW]
        rsd = rolling_second_derivative[i:i+WINDOW]
        rae = rolling_absolute_energy[i:i+WINDOW]
        
        o = remap(o, -1, 1)
        h = remap(h, -1, 1)
        l = remap(l, -1, 1)
        c = remap(c, -1, 1)
        v = remap(v, -1, 1)
        volat = remap(volat, -1, 1)
        rsd = remap(rsd, -1, 1)
        rae = remap(rae, -1, 1)
               
        y_i = (normal_close[i+WINDOW] - normal_open[i+WINDOW]) / normal_open[i+WINDOW]
        y_i_binary = 1 if y_i > 0 else 0
        x_i = np.column_stack((o, h, l, c, v, volat, rsd, rae))
    
    except Exception as e:
        print e
        break
        
    if np.isnan(x_i).any() or np.isnan(y_i).any():
        continue

    X.append(x_i)
    Y.append(y_i_binary)
    MARGINS.append(y_i)
    BENCHMARK.append(normal_close[i+WINDOW])
```

```python
def train_qda(X_train2, Y_train, X_test2, Y_test):
    clf = QuadraticDiscriminantAnalysis()
    clf.fit(X_train2, Y_train)

    pred = clf.predict(X_test2)
#     pred = [0 if p < 0.5 else 1 for p in pred]
    
    C = confusion_matrix(Y_test, pred)
    m = matthews_corrcoef(Y_test, pred)
    f = f1_score(Y_test, pred, average='micro')
    C = (C / C.astype(np.float).sum(axis=1))
    return m, C, f, pred
```

```python
def generate_dataset(start_day = 0, train_days = 30, val_days = 7, test_days = 14, flatten = True):
    TRAIN_START = 60*24*start_day
    TRAIN_END = TRAIN_START + int(60*24*train_days)
    VAL_START = TRAIN_END
    VAL_END = VAL_START + int(60*24*val_days)
    TEST_START = VAL_END
    TEST_END = TEST_START + int(60*24*test_days)

    X_train, X_val, X_test = X[TRAIN_START:TRAIN_END], X[VAL_START:VAL_END], X[TEST_START:TEST_END]
    Y_train, Y_val, Y_test = Y[TRAIN_START:TRAIN_END], Y[VAL_START:VAL_END], Y[TEST_START:TEST_END]
    M_train, M_val, M_test = MARGINS[TRAIN_START:TRAIN_END], MARGINS[VAL_START:VAL_END], MARGINS[TEST_START:TEST_END]
    B_train, B_val, B_test = BENCHMARK[TRAIN_START:TRAIN_END], BENCHMARK[VAL_START:VAL_END], BENCHMARK[TEST_START:TEST_END]

    if flatten:
        X_train2, X_val2, X_test2 = [x.flatten() for x in X_train], [x.flatten() for x in X_val], [x.flatten() for x in X_test]
    else:
        X_train2, X_val2, X_test2 = X_train, X_val, X_test
    X_train2, X_val2, X_test2 = np.array(X_train2), np.array(X_val2), np.array(X_test2)

#     plt.figure()
#     plt.plot(range(len(X_train2)), normal_close[TRAIN_START:TRAIN_END])
#     plt.plot(range(len(X_train2), len(X_train2)+len(X_val2)), normal_close[VAL_START:VAL_END])
#     plt.plot(range(len(X_train2)+len(X_val2), len(X_train2)+len(X_val2)+len(X_test2)), normal_close[TEST_START:TEST_END])
#     plt.show()

    return X_train2, Y_train, X_test2, Y_test, X_val2, Y_val, M_train, M_val, M_test, B_train, B_val, B_test
```

```python
def train_n_periods(START_DAY = 0, TRAIN_LENGTH = 30, VALIDATION_LENGTH = 7, TEST_LENGTH = 7, TEST_PERIODS = 10, plot = False):
    
    def get_profit(pred, margins):
        pred = np.array([-1 if y == 0 else 1 for y in pred])
        margins = np.array(margins)
        cs = np.cumsum([p * m for (p, m) in zip(pred, margins)])
        return cs
    
    Q_res = []
    F_scores = []
    cumsums = []
    benchmarks = []
    
    for i in range(START_DAY, START_DAY + TEST_LENGTH * TEST_PERIODS, TEST_LENGTH): 
        X_train2, Y_train, X_test2, \
        Y_test, X_val2, Y_val, \
        M_train, M_val, M_test, \
        B_train, B_val, B_test = generate_dataset(i, TRAIN_LENGTH, VALIDATION_LENGTH, TEST_LENGTH)
        
        m, c, f, pred = train_qda(X_train2, Y_train, X_test2, Y_test)
        profit = get_profit(pred, M_test)
        
        sharpe = np.mean(np.diff(profit)) / np.std(np.diff(profit))
        sharpe_benchmark = np.mean(np.diff(B_test)) / np.std(np.diff(B_test))
        
        if plot:
            plt.figure(figsize = (12, 6))
            plt.plot(profit, label = 'ALGO, ' + str(sharpe) + ', ' + str(m))
            plt.plot((B_test - B_test[0]) / np.max(B_test), label = 'ASSET, ' + str(sharpe_benchmark))
            plt.legend()
            plt.show()
        
        Q_res.append(sharpe - sharpe_benchmark)
        F_scores.append(f)
        cumsums.append(profit)
        benchmarks.append((B_test - B_test[0]) / np.max(B_test))
        
    return (np.mean(Q_res), np.std(Q_res), np.mean(F_scores), cumsums, benchmarks)
```

```python
def cartesian_hyperparameters_train_val_lengths(train_lengths, val_lengths, start = 0, trading_period = 7, test_periods = 5):
    res = []
    for i, tr_len in enumerate(train_lengths):
        resb = []
        for j, val_len in enumerate(val_lengths):
            try:
                (m, v, f, c, b) = train_n_periods(start, tr_len, val_len, trading_period, test_periods)
            except Exception as e:
                continue
            resb.append([m, v])
        res.append(resb)    
    res = np.array(res)
    return res


def get_min_max_points(res):
    max_profit = np.where(res[:, :, 0]==res[:, :, 0].max())
    i, j = max_profit[0][0], max_profit[1][0]
    least_volatility = np.where(res[:, :, 1]==res[:, :, 1].min())
    i2, j2 = least_volatility[0][0], least_volatility[1][0]
    return (i, j), (i2, j2)
```

```python
train_lengths = [14, 21]
val_lengths = [3, 7]

start_trading_day = 0
test_period = 7
optimization_periods = 4
trading_period = 1
all_periods = 50

ms, fs = [], []

for index in range(start_trading_day, all_periods * test_period, test_period):
 
    res = cartesian_hyperparameters_train_val_lengths(train_lengths, val_lengths, index, test_period, optimization_periods)

    (i, j), (i2, j2) = get_min_max_points(res)
    max_prof_tr, max_prof_val = train_lengths[i], val_lengths[j]
    min_vol_tr, min_vol_val = train_lengths[i2], val_lengths[j2]

    last_optimization_point = index + max(train_lengths) + max(val_lengths)
    training_start = last_optimization_point - (min_vol_tr + min_vol_val)
    traiding_start = training_start + min_vol_tr + min_vol_val
    
    print 'Optimization performed from %f' % (index)
    print 'Hyperparameters chose:', min_vol_tr, min_vol_val
    print 'TraiNing starts at day', training_start
    print 'TraiDing starts at day', traiding_start
    
    (m, v, f, c, b) = train_n_periods(training_start, min_vol_tr, min_vol_val, test_period, trading_period)
    
    print m, f
    ms.append(m)
    fs.append(f)
    
    plt.figure()
    plt.plot(c[0], label = 'Trades start day ' + str(traiding_start))
    plt.plot(b[0], label = 'Benchmark')
    plt.legend()
    plt.show()
```
